```python
from pandas.api.types import CategoricalDtype

df['production_date'] = pd.to_datetime(df['production_date'])
df['expiration_date'] = pd.to_datetime(df['expiration_date'])
median_date = df['expiration_date'].median()
df['expiration_date'].fillna(median_date, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
def fill_missing_weights_by_category(df, weight_column='weight', category_column='category'):

  category_means = df.groupby(category_column)[weight_column].transform(lambda x: round(x.mean(), 2))
  df[weight_column].fillna(category_means, inplace=True)

  return df
  df = fill_missing_weights_by_category(df)
df.head()
```

```python
df.isnull().sum()
```

```python
df['total_value'] = df['quantity'] * df['unit_price']
df.head()
```

```python
df['production_year'] = df['production_date'].dt.year
df.head()
```

```python
category_grouped = df.groupby('category').agg({'quantity': 'sum', 'total_value': 'sum'}).reset_index()
category_grouped.head()
```

```python
top_5_expensive_products = df.nlargest(5, 'unit_price')[['product_name', 'unit_price']].reset_index()
top_5_expensive_products.head()
```

```python
average_weight_by_manufacturer = df.groupby('manufacturer')['weight'].mean().round(2).reset_index()
average_weight_by_manufacturer.head()
```

```python
df.to_csv('cleaned_factory_data.csv', index=False)
```
