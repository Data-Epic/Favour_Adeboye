**User_query: When is the busiest time of day for the restaurant?**

FileMetadata:

file_name: "Waiters Cash_Sales -.csv"

file_path: "https://drive.google.com/file/d/1_RBqno4cu9kov98d2S13Ko9mPexp3gtM/view?usp=drive_link"

Capability: Simple dataset exploration

```python
import pandas as pd

# Set pandas display options for wider view
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Read the CSV file into a DataFrame
df = pd.read_csv("/content/sample_data/Waiters Cash_Sales -.csv")

# Display the first 5 rows
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names, non-null counts, and data types
print(df.info())
```

```python
df.head() # Display first 5 rows of the DataFrame
```

```python
df.describe() # Generate descriptive statistics of the DataFrame
```

```python
# Convert the 'Time' column to datetime format
df['Time'] = pd.to_datetime(df['Time'], errors='coerce')

# Extract the hour from the 'Time' column
df['Hour'] = df['Time'].dt.hour
df.iloc[10:15] # Display rows 10 to 14 (inclusive) of the DataFrame
```

```python
# Group the data by the 'Hour' column and sum the 'Total' sales
hourly_sales = df.groupby('Hour')[['Qty','Total']].sum().reset_index()

# Sort the results to find the busiest time
busiest_time = hourly_sales.sort_values(by=['Qty','Total'], ascending=False)

# Display the results
busiest_time.head(1) # Display the top row (busiest hour)
```

**User_query: What is their most expensive item?**

Capability: Follow On

```python
# Convert the 'Price/Unit' column to numeric datatype
df['Price/Unit'] = pd.to_numeric(df['Price/Unit'], errors='coerce')

# Find the maximum price
max_price = df['Price/Unit'].max()

# Filter the DataFrame to get all rows where the 'Price/Unit' equals the maximum price
most_expensive_items = df[df['Price/Unit'] == max_price]

# Display the result
most_expensive_items
```
